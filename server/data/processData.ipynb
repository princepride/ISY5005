```python
import csv

## 定义输入、表格和输出数据
#input_data = "What is the average duration of the project?"
#table_data = ["name", "start date", "end date", "duration"]
#output_data = "True"

## 定义CSV文件名和表头
#filename = "data.csv"
header = ["Input", "Table", "Output"]

data = []

with open(r"./data.txt", "r") as f:
    data = f.readlines()
# 打开CSV文件并将数据写入
with open(r"./data.csv", "w") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(header)
    for dataline in data:
        print(dataline)
        input_data = dataline.split(" Table: ")[0].split("Input: ")[1]
        table_data = dataline.split(" Table: ")[1].split("Output: ")[0]
        output_data = dataline.split("Output: ")[1].strip()
        writer.writerow([input_data, table_data, output_data])
```
