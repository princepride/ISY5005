```python
import requests
API_URL = "https://datasets-server.huggingface.co/first-rows?dataset=Aditya011%2Fautotrain-data-nl-to-sql&config=Aditya011--autotrain-data-nl-to-sql&split=train"
def query():
    response = requests.request("GET", API_URL)
    return response.json()
data = query()
```

```python
data
```

```python
import sqlite3

# 连接到数据库
conn = sqlite3.connect('database.db')
c = conn.cursor()

# 创建表
c.execute('''CREATE TABLE data
             (id INTEGER PRIMARY KEY,
             question TEXT,
             sql TEXT)''')

# 将行插入表中
for row in data['rows']:
    question = row['row']['question']
    sql = row['row']['sql']
    c.execute("INSERT INTO data (question, sql) VALUES (?, ?)", (question, sql))

# 提交更改并关闭连接
conn.commit()
conn.close()
```
